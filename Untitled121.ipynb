```python
import os
import json
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import load_model
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix

# ---------------------------------------------------------------------
# âš™ï¸ PATHS
# ---------------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Pathogen Mutation Predictor"
META_PATH = os.path.join(BASE_DIR, "archive", "metadata.csv")
MODEL_RF = os.path.join(BASE_DIR, "biomind_model.pkl")
MODEL_DL = os.path.join(BASE_DIR, "biomind_dl.h5")

OUT_CSV = os.path.join(BASE_DIR, "metadata-prediction.csv")
OUT_JSON = os.path.join(BASE_DIR, "metadata-prediction.json")

# ---------------------------------------------------------------------
# ðŸ§¬ 1ï¸âƒ£ LOAD METADATA
# ---------------------------------------------------------------------
print("[INFO] Loading metadata...")
data = pd.read_csv(META_PATH)
data.columns = data.columns.str.strip().str.replace(" ", "_").str.replace("-", "_").str.lower()
print(f"[INFO] Metadata shape: {data.shape}")

# Detect label column if any
target_candidates = [c for c in data.columns if any(x in c for x in ["resistance", "phenotype", "label", "response"])]
label_col = target_candidates[0] if target_candidates else None
if label_col:
    y_true = data[label_col].astype(str).str.lower().map({"resistant": 1, "susceptible": 0}).fillna(0).astype(int)
else:
    print("[WARN] No explicit target found, assuming unlabeled prediction.")
    y_true = np.zeros(len(data))

# ---------------------------------------------------------------------
# ðŸ§® 2ï¸âƒ£ SELECT NUMERIC FEATURES
# ---------------------------------------------------------------------
X = data.select_dtypes(include=[np.number, np.float64, np.int64]).fillna(0)
print(f"[INFO] Numeric columns detected: {X.shape[1]}")

# ---------------------------------------------------------------------
# ðŸ§  3ï¸âƒ£ LOAD MODELS
# ---------------------------------------------------------------------
print("[INFO] Loading trained models...")
rf_model = joblib.load(MODEL_RF)
dl_model = load_model(MODEL_DL)

# ---------------------------------------------------------------------
# ðŸ”§ 4ï¸âƒ£ FORCE ALIGN FEATURE COUNT
# ---------------------------------------------------------------------
expected_features = getattr(rf_model, "n_features_in_", X.shape[1])
current_features = X.shape[1]
print(f"[INFO] RandomForest expects {expected_features} features, current data has {current_features}.")

if current_features < expected_features:
    missing = expected_features - current_features
    for i in range(missing):
        X[f"dummy_feature_{i+1}"] = 0
    print(f"[FIX] Added {missing} zero-filled dummy features to match expected input.")

elif current_features > expected_features:
    # Trim extras (rare)
    X = X.iloc[:, :expected_features]
    print(f"[FIX] Trimmed extra columns to match expected {expected_features} features.")

print(f"[INFO] Final aligned input shape: {X.shape}")

# ---------------------------------------------------------------------
# ðŸ”¢ 5ï¸âƒ£ SCALE AND PREDICT
# ---------------------------------------------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("[INFO] Predicting...")
rf_pred = rf_model.predict(X_scaled)
dl_prob = dl_model.predict(X_scaled).flatten()
dl_pred = (dl_prob > 0.5).astype(int)

results_df = pd.DataFrame({
    "SampleID": np.arange(len(X)),
    "TrueLabel": y_true,
    "RF_Pred": rf_pred,
    "DL_Pred": dl_pred,
    "DL_Prob": np.round(dl_prob, 4)
})
print("[INFO] Predictions complete!")

# ---------------------------------------------------------------------
# ðŸ’¾ 6ï¸âƒ£ SAVE RESULTS
# ---------------------------------------------------------------------
results_df.to_csv(OUT_CSV, index=False)
results_df.to_json(OUT_JSON, orient="records", indent=4)
print(f"âœ… Results saved:\n - {OUT_CSV}\n - {OUT_JSON}")

# ---------------------------------------------------------------------
# ðŸ“Š 7ï¸âƒ£ DISPLAY SAMPLE RESULTS
# ---------------------------------------------------------------------
print("\n===== SAMPLE RESULTS =====")
print(results_df.head(10))

# ---------------------------------------------------------------------
# ðŸ“ˆ 8ï¸âƒ£ VISUALIZATION
# ---------------------------------------------------------------------
sns.set(style="whitegrid")

# Predicted Resistance Distribution
plt.figure(figsize=(6,4))
sns.countplot(x=dl_pred, palette="coolwarm")
plt.title("Predicted Resistance Distribution (Deep Learning)")
plt.xlabel("Predicted Label (0=Susceptible, 1=Resistant)")
plt.show()

# Probability Distribution
plt.figure(figsize=(6,4))
sns.histplot(dl_prob, bins=20, kde=True, color="orange")
plt.title("Deep Learning Prediction Confidence")
plt.xlabel("Predicted Probability of Resistance")
plt.show()

# Comparison Scatter
plt.figure(figsize=(6,6))
sns.scatterplot(x=rf_pred, y=dl_pred, hue=dl_prob, palette="viridis", s=70)
plt.title("RF vs DL Prediction Comparison")
plt.xlabel("RandomForest Prediction")
plt.ylabel("DeepLearning Prediction")
plt.show()

print("\nðŸŽ¯ Visualization complete. All results displayed and saved successfully.")
```
