```python
import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import load_model

# ---------------------------------------------------------------------
# âš™ï¸  Paths
# ---------------------------------------------------------------------
BASE_DIR = r"C:\Users\NXTWAVE\Downloads\Pathogen Mutation Predictor"
MODEL_RF = os.path.join(BASE_DIR, "biomind_model.pkl")
MODEL_DL = os.path.join(BASE_DIR, "biomind_dl.h5")
META_PATH = os.path.join(BASE_DIR, "archive", "metadata.csv")
DATA_PATHS = [
    os.path.join(BASE_DIR, "archive", "azm_sr_gwas_filtered_unitigs.Rtab"),
    os.path.join(BASE_DIR, "archive", "cfx_sr_gwas_filtered_unitigs.Rtab"),
    os.path.join(BASE_DIR, "archive", "cip_sr_gwas_filtered_unitigs.Rtab"),
]

OUT_PATH_JSON = os.path.join(BASE_DIR, "prediction_results.json")
OUT_PATH_CSV  = os.path.join(BASE_DIR, "prediction_results.csv")

# ---------------------------------------------------------------------
# ðŸ§© 1ï¸âƒ£ Load mutation data
# ---------------------------------------------------------------------
print("[INFO] Loading mutation data...")

dfs = []
for path in DATA_PATHS:
    df = pd.read_csv(path, sep="\t")
    df.reset_index(inplace=True)

    # Transpose if features-as-rows format
    if df.shape[1] > 5:
        df = df.T
        df.columns = df.iloc[0]
        df = df[1:]
    df.index.name = "SampleID"
    df.reset_index(inplace=True)
    dfs.append(df)

merged = dfs[0]
for i in range(1, len(dfs)):
    merged = pd.merge(merged, dfs[i], on="SampleID", how="outer")

merged["SampleID"] = merged["SampleID"].astype(str)
print(f"[INFO] Combined mutation shape: {merged.shape}")

# ---------------------------------------------------------------------
# ðŸ§« 2ï¸âƒ£ Load metadata
# ---------------------------------------------------------------------
meta = pd.read_csv(META_PATH)
meta.columns = meta.columns.str.strip().str.replace(" ", "_").str.replace("-", "_").str.lower()

possible_ids = ["sampleid", "sample_id", "id", "isolate", "run", "strain"]
meta_key = None
for key in possible_ids:
    if key in meta.columns:
        meta_key = key
        break

if meta_key is None:
    raise KeyError(f"No valid sample ID column found. Columns: {list(meta.columns)}")

meta.rename(columns={meta_key: "SampleID"}, inplace=True)
meta["SampleID"] = meta["SampleID"].astype(str)

data = pd.merge(meta, merged, on="SampleID", how="inner")
print(f"[INFO] Merged dataset: {data.shape}")

# ---------------------------------------------------------------------
# ðŸ§® 3ï¸âƒ£ Preprocess features
# ---------------------------------------------------------------------
target_candidates = [c for c in data.columns if any(x in c.lower() for x in ["resistance", "phenotype", "label", "response"])]
label_col = target_candidates[0] if target_candidates else None

if label_col:
    y_true = data[label_col].astype(str).str.lower().map({"resistant": 1, "susceptible": 0}).fillna(0).astype(int)
else:
    y_true = np.zeros(len(data))  # if no labels, placeholder

X = data.drop(columns=["SampleID"] + ([label_col] if label_col else []), errors="ignore")
X = X.select_dtypes(include=[np.number, np.float64, np.int64]).fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(f"[INFO] Feature matrix: {X_scaled.shape}")

# ---------------------------------------------------------------------
# ðŸ§  4ï¸âƒ£ Load models
# ---------------------------------------------------------------------
print("[INFO] Loading trained models...")
rf_model = joblib.load(MODEL_RF)
dl_model = load_model(MODEL_DL)

# ---------------------------------------------------------------------
# ðŸ”® 5ï¸âƒ£ Predict
# ---------------------------------------------------------------------
rf_pred = rf_model.predict(X_scaled)
dl_pred_prob = dl_model.predict(X_scaled).flatten()
dl_pred = (dl_pred_prob > 0.5).astype(int)

results_df = pd.DataFrame({
    "SampleID": data["SampleID"],
    "TrueLabel": y_true,
    "RF_Pred": rf_pred,
    "DL_Pred": dl_pred,
    "DL_Prob": np.round(dl_pred_prob, 4)
})
print("[INFO] Predictions complete.")

# ---------------------------------------------------------------------
# ðŸ’¾ 6ï¸âƒ£ Save outputs
# ---------------------------------------------------------------------
results_df.to_csv(OUT_PATH_CSV, index=False)
results_df.to_json(OUT_PATH_JSON, orient="records", indent=4)
print(f"âœ… Results saved to:\n {OUT_PATH_CSV}\n {OUT_PATH_JSON}")

# ---------------------------------------------------------------------
# ðŸ“Š 7ï¸âƒ£ Display Results On Screen
# ---------------------------------------------------------------------
print("\n===== SAMPLE PREDICTIONS =====")
print(results_df.head(10))

# ---------------------------------------------------------------------
# ðŸ“‰ 8ï¸âƒ£ Graphs and Visuals
# ---------------------------------------------------------------------
sns.set(style="whitegrid")

# Accuracy (if ground truth present)
if y_true.sum() > 0:
    acc_rf = np.mean(y_true == rf_pred)
    acc_dl = np.mean(y_true == dl_pred)
    plt.figure(figsize=(6,4))
    plt.bar(["RandomForest", "DeepLearning"], [acc_rf, acc_dl], color=["skyblue","orange"])
    plt.title("Model Accuracy Comparison")
    plt.ylabel("Accuracy")
    plt.show()

# Confusion Matrix
if y_true.sum() > 0:
    cm = pd.crosstab(y_true, dl_pred, rownames=["True"], colnames=["Predicted"])
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title("Confusion Matrix (DL)")
    plt.show()

# Distribution of Predictions
plt.figure(figsize=(6,4))
sns.countplot(x=dl_pred, palette="coolwarm")
plt.title("Predicted Resistance Distribution (DL)")
plt.xlabel("Predicted Label (0=Susceptible, 1=Resistant)")
plt.show()

# Probability Distribution
plt.figure(figsize=(6,4))
sns.histplot(dl_pred_prob, bins=20, kde=True, color="orange")
plt.title("DL Prediction Confidence Distribution")
plt.xlabel("Predicted Probability of Resistance")
plt.show()

# Comparison Scatter
plt.figure(figsize=(6,6))
sns.scatterplot(x=rf_pred, y=dl_pred, hue=dl_pred_prob, palette="viridis", s=70)
plt.title("RF vs DL Prediction Comparison")
plt.xlabel("RandomForest Prediction")
plt.ylabel("DeepLearning Prediction")
plt.show()
```
